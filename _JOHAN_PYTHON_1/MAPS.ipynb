```python
import gmaps
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from scipy.stats import linregress

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy


Johan_ironore_metadata = "JOhan All housing data.csv"
Johan_ironore_data = pd.read_csv(Johan_ironore_metadata)
Johan_ironoredf = pd.DataFrame(Johan_ironore_data)
Johan_ironoredf.head()
```

```python
locations = Johan_ironoredf[['LATITUDE', 'LONGITUDE']]
weights = Johan_ironoredf['PRICE']
PRINTMAP=locations.dropna(subset=['LATITUDE', 'LONGITUDE'])
fig = gmaps.figure()
fig.add_layer(gmaps.heatmap_layer(PRINTMAP, weights=None))
fig
```
